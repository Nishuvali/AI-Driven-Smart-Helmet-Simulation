```python
# Install required libraries
!pip install ultralytics transformers datasets opencv-python torch torchvision torchaudio matplotlib gdown geopy twilio scikit-learn tabulate

# Import necessary libraries
import torch
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
from ultralytics import YOLO
from transformers import ViTForImageClassification, ViTFeatureExtractor
from PIL import Image
from sklearn.metrics import precision_recall_fscore_support
from google.colab.patches import cv2_imshow
from geopy.geocoders import Nominatim
from twilio.rest import Client

# Twilio Credentials (Replace with your Twilio details)
TWILIO_SID = "ACf365ed9c4ee1d67a4b7a5ae9ecbfc674"
TWILIO_AUTH_TOKEN = "b78faf04cf138cb0073614c2b5cd87e0"
TWILIO_PHONE_NUMBER = "+12762626773"
EMERGENCY_CONTACT = "+919603349754"

# Function to send SMS alerts
def send_sms_alert(location_info):
    client = Client(TWILIO_SID, TWILIO_AUTH_TOKEN)
    message_body = f"🚨 Accident Detected! 🚨\nLocation: {location_info}"

    message = client.messages.create(
        body=message_body,
        from_=TWILIO_PHONE_NUMBER,
        to=EMERGENCY_CONTACT
    )

    print(f"📩 SMS Alert Sent! Message SID: {message.sid}")

# Function to get GPS coordinates (Simulated for Laptop)
def get_gps_coordinates():
    return 37.7749, -122.4194  # Example: San Francisco (Replace with real GPS if available)

# Function to convert GPS coordinates to an address
def get_location_info(lat, lon):
    geolocator = Nominatim(user_agent="smart-helmet")
    location = geolocator.reverse((lat, lon), exactly_one=True)
    return location.address if location else "Unknown Location"

# Load Pre-trained YOLOv8 Model
yolo_model = YOLO('yolov8n.pt')

# Load Pre-trained Vision Transformer (ViT) Model
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
vit_model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')

# Download sample test video
!gdown --id 1lOF4KRRnOU6koGRD5nSh_RS1HK3dTy8P -O sample_test_video.mp4

# Video Processing
video_path = "sample_test_video.mp4"
cap = cv2.VideoCapture(video_path)
frame_count = 0
max_frames = 10
processing_times = []
true_labels = []
predicted_labels = []

while cap.isOpened() and frame_count < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    start_time = time.time()

    if frame_count % 5 == 0:  # Process every 5th frame
        detection_results = yolo_model(frame)
        accident_detected = False

        for result in detection_results:
            for box in result.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                label = f"{result.names[int(box.cls[0])]}: {box.conf[0]:.2f}"
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

                if "accident" in label:
                    accident_detected = True  # Mark as accident detected

        predicted_labels.append(1 if accident_detected else 0)
        true_labels.append(1 if frame_count in [15, 30, 45] else 0)  # Simulating accident ground truth

        # If an accident is detected, send an emergency alert
        if accident_detected:
            lat, lon = get_gps_coordinates()
            location_info = get_location_info(lat, lon)
            print(f"🚨 Accident Detected at {location_info}")
            send_sms_alert(location_info)

        cv2_imshow(frame)

    processing_times.append(time.time() - start_time)

cap.release()

# Calculate Model Performance Metrics
precision, recall, f1, _ = precision_recall_fscore_support(true_labels, predicted_labels, average='binary')

print(f"\n🚀 **Performance Metrics**")
print(f"Precision: {precision:.2f}, Recall: {recall:.2f}, F1-score: {f1:.2f}")
print(f"Average Processing Time per Frame: {np.mean(processing_times):.4f} seconds")

# Display Metrics in a Table
from tabulate import tabulate
results_table = [
    ["Precision", f"{precision:.2f}"],
    ["Recall", f"{recall:.2f}"],
    ["F1-score", f"{f1:.2f}"],
    ["Avg Processing Time per Frame", f"{np.mean(processing_times):.4f} sec"]
]

print(tabulate(results_table, headers=["Metric", "Value"], tablefmt="grid"))
```
